```python
# Importando bibliotecas python

from pyspark.sql import functions as f
```

```python
df = spark.table('silver.crypto_exchange')
df.limit(5).display()
```

```python
# Check 1: Verificando se há preços negativos
invalid_price = df.filter(f.col('exchange_rate') < 0).count()
#print(invalid_price) 

# Check 2: Verificando se há cryptos sem código
null_crypto = df.filter(f.col('crypto_code').isNull()).count()

## Check 3: Verificando se há spread negativo 
##invalid_spreda = df.filter(f.col('spread') < 0).count()
```

```python
# Forçando erro no pipeline quando encontrar alguma das validações acima 

if invalid_price > 0:
    raise Exception("Data Quality Error: Exchange Rate Negativo")

if null_crypto > 0:
    raise Exception("Data Quality Error: Crypto Code Nulo")
```
